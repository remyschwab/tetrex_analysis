# Analysis of kmer Complexity

```python
import os
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

## For the RegEx LMAEGLYN.SVRVRSDI

```python
input_path = "/Users/rschwab/Desktop/simple_complexity_example.txt"
```

```python
def parse_complexity(window):
    no_return = window.strip()
    no_lbracket = no_return.strip("[")
    no_brackets = no_lbracket.strip("]")
    spalt = no_brackets.split(",")
    cleaned = []
    for c in spalt:
        if c.isnumeric():
            cleaned.append(int(c))
    return cleaned
```

```python
cmplx_matrix = []
with open(input_path) as handle:
    for i, line in enumerate(handle.readlines()):
        line_info = line.split(":")
        cmplx_matrix.append((line_info[0], parse_complexity(line_info[1])))
```

```python
figure(figsize=(20, 10))
all = []
for window in cmplx_matrix:
        all.append(sum(window[1]))
plt.plot(all)
plt.axvline(x=9, color="red")
plt.axvline(x=66, color="red")
```

```python
characters = []
for window in cmplx_matrix:
    if window[0].isalpha():
        characters.append(sum(window[1]))
    else:
        characters.append(None)
# plt.scatter(range(191), characters)
plt.plot(characters)
plt.axvline(x=9, color="red")
plt.axvline(x=66, color="red")
```

```python
splits = []
for window in cmplx_matrix:
        if window[0] == "":
                splits.append(sum(window[1]))
plt.plot(splits)
plt.axvline(x=9, color="red")
plt.axvline(x=66, color="red")
```

```python
ghosts = []
for window in cmplx_matrix:
        if window[0] == "":
                ghosts.append(sum(window[1]))
plt.plot(ghosts)
```
